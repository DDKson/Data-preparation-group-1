```python
mask = np.triu(np.ones_like(application_train.corr()))
sns.heatmap(application_train.corr(), mask = mask)
```

```python
# Create correlation matrix
corr_matrix = application_train.corr().abs()

# Select upper triangle of correlation matrix
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.85
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.85)]
to_drop
```

```python
# Drop features 
application_train.drop(to_drop, axis=1, inplace=True)
application_train.shape
```

```python
''' Detection '''
# IQR
Q1 = np.percentile(application_train["CNT_CHILDREN"], 25, interpolation = 'midpoint')
Q3 = np.percentile(application_train["CNT_CHILDREN"], 75, interpolation = 'midpoint')

IQR = Q3 - Q1
 
print("Old Shape: ", application_train.shape)
 
# Upper bound
upper = np.where(application_train["CNT_CHILDREN"] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(application_train["CNT_CHILDREN"] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
application_train.drop(upper[0], inplace = True)
application_train.drop(lower[0], inplace = True)
 
print("New Shape: ", application_train.shape)
```
